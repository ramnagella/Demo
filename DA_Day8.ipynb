```
#innerjoin()
import pandas as pd
Employees=pd.DataFrame({
    'Emp_ID':[1,2,3,4],
    'Name':['Ram','Shatha','Vamsi','Kiran'],
})
Dept=pd.DataFrame({
    'Emp_ID':[2,3,4,5],
    'Dept':['IT','HR','IT','HR'],
})
pd.merge(Employees,Dept,how='inner',on='Emp_ID')
```

```
pd.merge(Employees,Dept)
#will give same o/p as above because it will default unique column key i.e semp_id and inner join
```

```
pd.merge(Employees,Dept,how='left',on='Emp_ID')
```

```
pd.merge(Employees,Dept,how='right',on='Emp_ID')
```

```
#merging multiple keys--in Pandas we cant give alias names for the keys like product_ id must be same in botht DF's
#it is possible in SQL
orders=pd.DataFrame({
    'Order_ID':[1,2,3],
    'Product_ID':[101,102,103],
    'Store':['A','B','A']
})
inventory=pd.DataFrame({
    'Product_ID':[101,102,103,104],
    'Store':['A','A','B','A'],
    'Quantity':[10,5,20,15]
})
pd.merge(orders,inventory,how='inner',on=['Product_ID','Store'])
```

```
products=pd.DataFrame({
  'Product_ID':[101,102],
  'Prod_name':['Laptop','Tablet']
})
Supplier=pd.DataFrame({
    'Supplier_ID':[1,2],
    'Product_ID':[101,102],
    'Prod_name':['Laptop Pro','Tablet Plus']
})
pd.merge(products,Supplier,on='Product_ID')
#Prod_name_x and Prod_name_y were automatically created _x is for first DF and _y for second DF
```

```
pd.merge(products,Supplier,on='Product_ID',suffixes=('_product','_supplier'))
```

```
#Advance merging techniques --
import pandas as pd
Employees=pd.DataFrame({
    'Emp_ID':[1,2,3,4],
    'Name':['Ram','Shatha','Vamsi','Kiran']
})
Projects=pd.DataFrame({
    'Emp_ID':[2,3,5],
    'Project_Name':['Alpha','Beta','Gama']
})
pd.merge(Employees,Projects,on='Emp_ID',how='inner')
```

```
import pandas as pd
Employees=pd.DataFrame({
    'Emp_ID':[1,2,3,4],
    'Name':['Ram','Shatha','Vamsi','Kiran']
})
Projects=pd.DataFrame({
    'Emp_ID':[2,3,5],
    'Project_Name':['Alpha','Beta','Gama']
})
pd.merge(Employees,Projects,on='Emp_ID',how='outer')
#Common values will get names and other non-common values will get NaN.Will get entire thing
```

```
import pandas as pd
Employees=pd.DataFrame({
    'Emp_ID':[1,2,3,4],
    'Name':['Ram','Shatha','Vamsi','Kiran']
})
Projects=pd.DataFrame({
    'Emp_ID':[2,3,5],
    'Project_Name':['Alpha','Beta','Gama']
})
pd.merge(Employees,Projects,on='Emp_ID',how='outer',indicator=True)
#Default indicator is False
```

```
#Ensuring one to one relationship using validate attribute
orders=pd.DataFrame({
    'Order_ID':[1,2,3],
    'Customer_ID':[101,102,103]
})
Shipments=pd.DataFrame({
    'Shipment_ID':[1001,1002,1003],
    'Order_ID':[1,2,3]
})
pd.merge(orders,Shipments,on='Order_ID')
```

```
orders=pd.DataFrame({
    'Order_ID':[1,2,3],
    'Customer_ID':[101,102,103]
})
Shipments=pd.DataFrame({
    'Shipment_ID':[1001,1002,1003],
    'Order_ID':[1,2,3]
})
pd.merge(orders,Shipments,on='Order_ID',validate='one_to_one')
#both will give same answer one_to_one means each order corresponds to exactly one shipment,no dupicates and missing values
#otherwise will get not a number
```
